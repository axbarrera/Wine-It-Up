```python
# Dependencies 
import pandas as pd
import numpy as np
```

```python
#File to Load
wine_file = "data/winemag-data-130k-v2.csv"
```

```python
wine_df = pd.read_csv(wine_file)
```

```python
wine_df.head()
```

```python
wine_df.columns = ['id','country','description','designation','points','price', 'province', 'first_region', 'second_region', 'taster_name','taster_twitter_handle', 'title', 'variety', 'winery']
```

```python
wine_df.head()
```

```python
wine_df = wine_df.reset_index(drop=True)
```

```python
wine_df.head()
```

```python
wine_fulldata = wine_df.dropna()
```

```python
wine_fulldata.head()
```
